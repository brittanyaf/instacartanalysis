```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge.head(100)
```

```python
ords_prods_merge.groupby('loyalty_flag')['prices'].mean()
```

Based on the above information, New customers spend more on average than Regular or Loyal customers. Loyal customers spend the lowest of all of the loyalty flags.

```python
ords_prods_merge['max_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.max)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_price'] > 10, 'spending_flag'] = 'High Spender'
```

```python
ords_prods_merge.head(100)
```

```python
ords_prods_merge['med_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['med_frequency'] > 20, 'order_freq_flag'] = 'Non-frequent Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['med_frequency'] <= 20) & (ords_prods_merge['med_frequency'] > 10), 'order_freq_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['med_frequency'] <= 10, 'order_freq_flag'] = 'Frequent Customer'
```

```python
#Creating Regions Column
result = []

for value in final_data['State']:
    
    if value == 'Maine' or value == 'New Hampshire' or value == 'Vermont' or value == 'Massachusetts' or value == 'Rhode Island' or value == 'Connecticut' or value == 'New York' or value == 'New Jersey' or value == 'Pennsylvania':
        result.append('Northeast')
    elif value == 'Wisconsin' or value == 'Michigan' or value == 'Illinois' or value == 'Indiana' or value == 'Ohio' or value == 'North Dakota' or value == 'South Dakota' or value == 'Nebraska' or value == 'Kansas' or value == 'Minnesota' or value == 'Iowa' or value == 'Missouri' :
        result.append('Midwest')
    elif value == 'Delaware' or value == 'Maryland' or value == 'District of Columbia' or value == 'Virginia' or value == 'West Virginia' or value == 'North Carolina' or value == 'South Carolina' or value == 'Georgia' or value == 'Florida' or value == 'Kentucky' or value == 'Tennessee' or value == 'Mississippi' or value == 'Alabama' or value == 'Oklahoma' or value == 'Texas' or value == 'Arkansas' or value == 'Louisiana':
        result.append('South')
    else:
        result.append('West')
```

```python
final_data.head()
```

```python
pd.set_option('display.max_columns', None)
final_data.head()
```

```python
#Adding new column to datafram
final_data['Region'] = result
```

```python
#Confirming Region Column was Added
pd.set_option('display.max_columns', None)
final_data.head()
```

```python
#Checking for Spending Differences Among Regions
final_data.groupby('Region')['spending_flag'].value_counts()
```

Per the above value counts per region, it appears the South has the most high spenders, followed by the West, Midwest, and Northeast.  There is not much difference between West and Midwest.

```python
small.shape
```

```python
# Creating Customer Profiles
# Young Single Adults 
small.loc[(small['fam_status'] == 'single') & (small['Age'] <= 30), 'cust_profile'] = 'Young_single'
```

```python
# Young Married Adults with Children
small.loc[(small['fam_status'] == 'married') & (small['Age'] <= 30) & (small['no_of_dependents'] > 0), 'cust_profile'] = 'Young_family'
```

```python
#Young Married Adults without Children
small.loc[(small['fam_status'] == 'married') & (small['Age'] <= 30) & (small['no_of_dependents'] == 0), 'cust_profile'] = 'Young_family_no_kids'
```

```python
#All Married Adults
small.loc[(small['fam_status'] == 'married'), 'cust_profile'] = 'Married_adults'
```

```python
#All Married Adults with Children
small.loc[(small['fam_status'] == 'married') & (small['no_of_dependents'] > 0), 'cust_profile'] = 'Married_with_children_all_ages'
```

```python
#All Married Adults without Children
small.loc[(small['fam_status'] == 'married') & (small['no_of_dependents'] == 0), 'cust_profile'] = 'Married_no_children_all_ages'
```

```python
#Single Adults (Including Divorced/Widowed)
small.loc[(small['fam_status'] == 'married') & (small['fam_status'] == 'divorced/widowed'), 'cust_profile'] = 'Single_adults'
```

```python
#Adding "Other" for Profiles that Don't Match Above
small['cust_profile'] = small['cust_profile'].fillna('Other')
```

```python
#Checking Value Counts of Each Profile
small['cust_profile'].value_counts(dropna = False)
```

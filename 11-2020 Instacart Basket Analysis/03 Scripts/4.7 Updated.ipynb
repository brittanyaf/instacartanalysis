```python
#Define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
all_data_merged.loc[all_data_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
all_data_merged.loc[(all_data_merged['prices'] <= 15) & (all_data_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
all_data_merged.loc[all_data_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Creating Most Busiest Days column
result = []

for value in all_data_merged["orders_day_of_week"]:
  if value >= 1:
    result.append("Busiest days")
  elif value == 3 or 4:
    result.append("Least busy days")
  else:
    result.append("Regularly busy")
```

```python
all_data_merged['busiest_days'] = result
```

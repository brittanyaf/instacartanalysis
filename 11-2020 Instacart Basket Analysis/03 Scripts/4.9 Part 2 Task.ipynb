```python
#Creating Histogram for Order Hour of Day
hist_hod = final_data['order_hour_of_day'].plot.hist(bins = 70)
```

On the Y axis, we see the frequency of orders, and on the X axis, we see the hour of orders (0-24).  This histogram shows us that the most orders are placed between hours 10-15 (10am-3pm) with the highest number of orders during the 10am hour.

```python
#Creating Bar Chart for Customer Loyalty
loyalty_bar = final_data['loyalty_flag'].value_counts().plot.bar()
```

The above bar chart indicates that Regular Customers place the most orders of the three loyalty flags.

```python
#Creating Line Chart from Sample for Hour of Day/Prices
price_line = small[['order_hour_of_day','prices']]
```

```python
price_line_chart = sns.lineplot(data = price_line, x = 'order_hour_of_day',y = 'prices')
```

There is a difference in expenditure depending on hour of day, with higher expenditures in the early hours of the day.

```python
#Creating Line Chart from Sample for Age/Family Status/Number of Dependents
cust_age_dep = small[['Age','no_of_dependents']]
```

```python
cust_age_dep_line = sns.lineplot(data = cust_age_dep, x = 'Age',y = 'no_of_dependents')
```

The chart looks erratic, but since the number of dependents ranges from 0-3, there is a small difference.  It looks like most customers have between 1-2 dependents, per this chart, and age does not play a large difference in number of dependents.
